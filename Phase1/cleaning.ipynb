### 1. Handle internet_service Missing Values

```python
# Fill internet_service null values with 'No Internet Service'
df['internet_service'] = df['internet_service'].fillna('No Internet Service')
print("internet_service missing values after filling:", df['internet_service'].isnull().sum())
```

### 2. Check for Duplicates

```python
# Check for duplicate customer_ids
duplicate_count = df['customer_id'].duplicated().sum()
print(f"Number of duplicate customer_ids: {duplicate_count}")

if duplicate_count > 0:
    print("\nDuplicate customer_ids:")
    print(df[df['customer_id'].duplicated(keep=False)]['customer_id'].unique())
else:
    print("No duplicate customer_ids found.")
```

### 3. Validate and Convert Data Types

```python
# Check current data types
print("Current data types:")
print(df.dtypes)
print("\n" + "="*50 + "\n")

# Convert date columns to datetime
date_columns = ['account_created_date', 'churn_date']
for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
        print(f"Converted {col} to datetime")

print("\n" + "="*50 + "\n")
print("Data types after conversion:")
print(df[date_columns].dtypes)
```

### 4. Check for Outliers in Numerical Columns

```python
# Check for outliers in key numerical columns
numerical_cols = ['age', 'monthly_charges_thb', 'total_charges_thb', 'tenure_months', 
                  'data_usage_gb', 'call_minutes', 'num_support_tickets']

print("Statistical Summary of Numerical Columns:")
print("="*80)
print(df[numerical_cols].describe())

print("\n" + "="*80 + "\n")

# Check for negative values that shouldn't be negative
for col in numerical_cols:
    negative_count = (df[col] < 0).sum()
    if negative_count > 0:
        print(f"WARNING: {col} has {negative_count} negative values")
        
# Check for unrealistic values
print("\nChecking for unrealistic values:")
print(f"Age > 100: {(df['age'] > 100).sum()} rows")
print(f"Age < 18: {(df['age'] < 18).sum()} rows")
print(f"Tenure months > 120 (10 years): {(df['tenure_months'] > 120).sum()} rows")
print(f"Monthly charges = 0: {(df['monthly_charges_thb'] == 0).sum()} rows")
print(f"Total charges = 0: {(df['total_charges_thb'] == 0).sum()} rows")
```

### 5. Standardize Categorical Values

```python
# Check categorical columns for inconsistent values
categorical_cols = ['gender', 'partner', 'dependents', 'contract_type', 'payment_method', 
                   'paperless_billing', 'online_security', 'online_backup', 'device_protection',
                   'tech_support', 'multiple_lines', 'international_roaming',
                   'plan_downgrade_6months', 'plan_upgrade_6months']

print("Unique values in categorical columns:")
print("="*80)
for col in categorical_cols:
    if col in df.columns:
        unique_vals = df[col].unique()
        print(f"\n{col}: {unique_vals}")
        print(f"  Count: {len(unique_vals)} unique values")
```

### 6. Final Data Quality Check

```python
# Final summary of data quality
print("FINAL DATA QUALITY REPORT")
print("="*80)
print(f"\nTotal rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")

# Missing values summary
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percent': missing_percent
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
print("\nRemaining Missing Values:")
if len(missing_df) > 0:
    print(missing_df)
else:
    print("No missing values!")

# Data shape
print(f"\nDataset shape: {df.shape}")
print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print("\n" + "="*80)
print("Data cleaning completed successfully!")
```
